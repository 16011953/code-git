```python
#CODE FOR QUALITATIVE DATA ANALYSIS(THEMATIC ANALYSIS)
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

# Sample interview responses
responses = [
    "The primary security challenge is terrorism, particularly from non-state actors like ISIS, which destabilizes the region.",
    "The Arab League and GCC have not been effective due to political divisions.",
    "Countries like Saudi Arabia, Iran, and Turkey are driven by national interests, which often clash.",
    "Cybersecurity should be central in a new security framework."
]

# Preprocessing: Tokenization and removing stopwords
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = word_tokenize(text.lower())
    filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    return filtered_tokens

# Apply preprocessing to each response
processed_responses = []
for response in responses:
    processed_responses.extend(preprocess(response))

# Thematic analysis: Count word frequency
word_freq = Counter(processed_responses)

# Display the most common themes (words)
print(word_freq.most_common(10))
```

```python
#CODE FOR THEMATIC AND COMPARATIVE ANALYSIS
from collections import defaultdict

# Sample interview responses from multiple stakeholders
responses_stakeholders = {
    "Diplomat 1": "Terrorism is the greatest challenge. The Arab League is ineffective due to political divisions.",
    "Security Expert 1": "Cybersecurity is emerging as a major issue, but terrorism remains the primary threat.",
    "Diplomat 2": "Regional cooperation is weak, especially due to national interests taking precedence.",
    "Security Analyst 1": "The GCC and Arab League have not been effective due to mistrust among nations."
}

# Function to preprocess and tokenize responses
def preprocess(text):
    tokens = word_tokenize(text.lower())
    return [word for word in tokens if word.isalpha() and word not in stopwords.words('english')]

# Store themes from each stakeholder
themes = defaultdict(list)

for stakeholder, response in responses_stakeholders.items():
    themes[stakeholder].extend(preprocess(response))

# Compare thematic similarities and differences
for stakeholder, theme_list in themes.items():
    print(f"{stakeholder}: {Counter(theme_list).most_common(5)}")

# Comparative analysis: Compare themes between stakeholders
common_themes = set(themes["Diplomat 1"]).intersection(set(themes["Security Expert 1"]))
print(f"\nCommon themes between Diplomat 1 and Security Expert 1: {common_themes}")
```

```python
import matplotlib.pyplot as plt

# Threat categories and initial levels
threats = ['Terrorism', 'Geopolitical Rivalries', 'Cybersecurity Threats']
initial_threat_levels = [80, 70, 60]  # Threat levels (on a scale of 0 to 100)

# Interventions and their effectiveness in percentage
interventions = ['Counterterrorism', 'Diplomacy', 'Cybersecurity']
effectiveness = [30, 40, 50]  # How much each intervention reduces the threat

# Calculate new threat levels after applying interventions
updated_threat_levels = [initial - (initial * eff / 100) for initial, eff in zip(initial_threat_levels, effectiveness)]

# Plotting the before and after threat levels using bar charts
plt.figure(figsize=(8, 5))

# Initial threat levels
plt.bar(threats, initial_threat_levels, color='red', width=0.4, label='Initial Threat Levels', edgecolor='black')

# New threat levels after interventions
plt.bar(threats, updated_threat_levels, color='green', width=0.4, label='Threat Levels After Interventions', edgecolor='black')

# Add labels and title
plt.title('Impact of Interventions on Security Threats', fontsize=14)
plt.xlabel('Security Threats', fontsize=12)
plt.ylabel('Threat Levels (0-100)', fontsize=12)
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
